#### Total US Area

```python
pipeline = [
    {"$group" : {"_id" : 0,
                 "totalArea" :{"$sum" : "$areaM"},
                 "avgArea" : {"$avg" : "$areaM"}}}
]
aggResult = collection.aggregate(pipeline)
```

```python
pd.DataFrame(list(aggResult))
```

```python
db.command('aggregate', 'consensus', pipeline=pipeline, explain=True)
```

#### Area by Region

```python
pipeline = [
    {"$group" : {"_id" : "$region",
                "totalArea" : {"$sum" : "$areaM"},
                "avgArea" : {"$avg" : "$areaM"},
                "numStates" : {"$sum" : 1},
                "states" : {"$push" : "$name"}}}
]
aggResult = collection.aggregate(pipeline)
```

```python
pd.DataFrame(list(aggResult))
```

#### Total US population by year

```python
pipeline = [
    {"$unwind" : "$data"},
    {"$group" : {"_id" : "$data.year",
                "totalPop" : {"$sum" : "$data.totalPop"}}},
    {"$sort" : {"totalPop" : -1}}
]
aggResult = collection.aggregate(pipeline)
```

```python
pd.DataFrame(list(aggResult))
```

#### Southern State Population by year

```python
pipeline = [
    {"$match" : {"region" : "South"}},
    {"$unwind" : "$data"},
    {"$group" : {"_id" : "$data.year",
                "totalPop" : {"$sum" : "$data.totalPop"}}}
]
aggResult = collection.aggregate(pipeline)
```

```python
pd.DataFrame(list(aggResult))
```

#### Delta Population by state between 1990 and 2010

```python
pipeline = [
    {"$unwind" : "$data"},
    {"$sort" : {"data.year" : 1}},
    {"$group" : {"_id" : "$name",
                "pop1990" : {"$first" : "$data.totalPop"},
                "pop2010" : {"$last" : "$data.totalPop"}}},
    {"$project" : {"_id" : 0,
                  "name" : "$_id",
                  "delta" : {"$subtract" : ["$pop2010", "$pop1990"]},
                  "pop1990" : 1,
                  "pop2010" : 1}}
]
aggResult = collection.aggregate(pipeline)
```

```python
pd.DataFrame(list(aggResult))
```

#### compare the number of people living within 500km of Memphis [90, 35], TN in 1990, 2000, 2010

```python
pipeline = [
    {"$geoNear" : {
        "near" : {"type" : "Point", "coordinates" : [90, 35]},
        "distanceField" : "dist.calculated",
        "maxDistance" : 500000,
        "includeLocs" : "dist.location",
        "spherical" : True
    }},
    {"$unwind" : "$data"},
    {"$group" : {"_id" : "$data.year",
                "totalPop" : {"$sum" : "$data.totalPop"},
                "states" : {"$addToSet" : "$name"}}},
    {"$sort" : {"_id" : 1}}
]
aggResult = collection.aggregate(pipeline)
```

```python
df = pd.DataFrame(list(aggResult))
df.style.set_properties(subset=['states'], **{'width': '500px'})
```

#### save results into a Collection

```python
pipeline = [
    {"$geoNear" : {
        "near" : {"type" : "Point", "coordinates" : [90, 35]},
        "distanceField" : "dist.calculated",
        "maxDistance" : 500000,
        "includeLocs" : "dist.location",
        "spherical" : True
    }},
    {"$unwind" : "$data"},
    {"$group" : {"_id" : "$data.year",
                "totalPop" : {"$sum" : "$data.totalPop"},
                "states" : {"$addToSet" : "$name"}}},
    {"$sort" : {"_id" : 1}},
    {"$out" : "peopleNearMemphis"}
]
aggResult = collection.aggregate(pipeline)
```

#### Which US Division has the fastest growing population density?

```python
pipeline = [
    {"$match" : {"data.totalPop" : {"$gt" : 1000000}}},
    {"$unwind" : "$data"},
    {"$sort" : {"data.year" : 1}},
    {"$group" : {"_id" : "$name",
                "pop1990" : {"$first" : "$data.totalPop"},
                "pop2010" : {"$last" : "$data.totalPop"},
                "areaM" : {"$first" : "$areaM"},
                "division" : {"$first" : "$division"}}},
    {"$group" : {"_id" : "$division",
                "totalPop1990" : {"$sum" : "$pop1990"},
                "totalPop2010" : {"$sum" : "$pop2010"},
                "totalAreaM" : {"$sum" : "$areaM"}}},
    {"$match" : {"totalAreaM" : {"$gt" : 100000}}},
    {"$project" : {"_id" : 0,
                  "dividion" : "$_id",
                  "density1990" : {"$divide" : ["$totalPop1990", "$totalAreaM"]},
                  "density2010" : {"$divide" : ["$totalPop2010", "$totalAreaM"]},
                  "denDelta" : {"$subtract" : [{"$divide" : ["$totalPop2010", "$totalAreaM"]},
                                              {"$divide" : ["$totalPop1990", "$totalAreaM"]}]},
                  "totalAreaM" : 1,
                  "totalPop1990" : 1,
                  "totalPop2010" : 1}},
    {"$sort" : {"denDelta" : -1}},
    {"$limit" : 1}
]
aggResult = collection.aggregate(pipeline)
```

```python
pd.DataFrame(list(aggResult))
```

```python
db.command('aggregate', 'consensus', pipeline=pipeline, explain=True)
```
